```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

def ver(Ω_em_pi=0.3):
    Ω = Ω_em_pi * np.pi
    n = np.arange(0, 30)
    x_discreto = np.cos(Ω * n)

    # Sinal contínuo
    t = np.linspace(0, 29, 500)
    x_cont = np.cos(Ω * t)

    plt.figure(figsize=(20, 6))

    plt.plot(t, x_cont, 'r-', alpha=0.6, linewidth=2, label='sinal contínuo')
    plt.stem(n, x_discreto, label='amostras x[n]')
    plt.plot(n, x_discreto, 'b-', alpha=0.4)

    plt.title(f'x[n] = cos(Ω n)    →    Ω = {Ω_em_pi:.3f} π rad/amostra', fontsize=16)
    plt.xlabel('n (amostras)')
    plt.ylabel('amplitude')
    plt.xticks(np.arange(0, 30))
    plt.ylim(-1.15, 1.15)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

interact(ver,
         Ω_em_pi=widgets.FloatSlider(min=0, max=2.0, step=0.05, value=0.3,
                                     description='Ω / π',
                                     style={'description_width': 'initial'}));
```
